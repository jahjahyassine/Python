```python
import random
import time

def partition_lumoto(numbers, start, end):
    pivot = numbers[end]
    i = start - 1
    for j in range(start, end):
        if numbers[j] <= pivot:
            i += 1
            numbers[i], numbers[j] = numbers[j], numbers[i]
    numbers[i + 1], numbers[end] = numbers[end], numbers[i + 1]
    return i + 1

def partition2(numbers, start, end):
    pivot = numbers[end]
    index_inf = start
    index_sup = end - 1
    while index_inf <= index_sup:
        if numbers[index_inf] <= pivot:
            index_inf += 1
        else:
            numbers[index_inf], numbers[index_sup] = numbers[index_sup], numbers[index_inf]
            index_sup -= 1
    numbers[index_inf], numbers[end] = numbers[end], numbers[index_inf]
    return index_inf

def triRapide_Lumoto(numbers, start, end):
    if start < end:
        p = partition_lumoto(numbers, start, end)
        triRapide_Lumoto(numbers, start, p - 1)
        triRapide_Lumoto(numbers, p + 1, end)

def triRapide_partition2(numbers, start, end):
    if start < end:
        p = partition2(numbers, start, end)
        triRapide_partition2(numbers, start, p - 1)
        triRapide_partition2(numbers, p + 1, end)

numbers = [random.randint(1, 100000) for x in range(100000)]
numbers2 = numbers.copy()

# ---- Lomoto ----

time_start = time.time()
triRapide_Lumoto(numbers, 0, len(numbers) - 1)
time_end = time.time()
time_execution = time_end - time_start
print(f"Lomuto total time = {time_execution:.5f}s\n")

with open("time(Lumoto).txt", "a") as file:
    file.write(str(time_execution) + "\n")

# ---- Partition2 ----

time_start2 = time.time()
triRapide_partition2(numbers2, 0, len(numbers2) - 1)
time_end2 = time.time()
time_execution2 = time_end2 - time_start2
print(f"Partition2 total time = {time_execution2:.5f}s\n")

with open("time(Partition2).txt", "a") as file:
    file.write(str(time_execution2) + "\n")
```
