```python
#Tri par Insertion

import time
import random

list_numbers = [random.randint(1, 10500) for x in range(10000)]

print("The list before: ")
print(list_numbers)

time_start = time.time()

for i in range(1, len(list_numbers)):
    for j in range(i):
        if list_numbers[i] < list_numbers[j]:
            num_temp = list_numbers[i]
            del list_numbers[i]
            list_numbers.insert(j, num_temp)
            break
            
time_end = time.time()
with open("time(Insertion).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print(f"The total time is {time_end - time_start}\n")
print("The list after: ")
print(list_numbers)
```

```python
#Tri par Selection

import time
import random

list_numbers = [random.randint(1, 55000) for x in range(10000)]

print("The list before: ")
print(list_numbers)

time_start = time.time()

for i in range(0, len(list_numbers) - 1):
    num_min = list_numbers[i]
    for j in range(i + 1, len(list_numbers)):
        if num_min > list_numbers[j]:
            num_min = list_numbers[j]
            index_min = j
    list_numbers[i], list_numbers[index_min] = list_numbers[index_min], list_numbers[i]

time_end = time.time()
with open("time(Selection).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print("The list after: ")
print(list_numbers)
```

```python
#Recherche dichotomique

import time
import random

list_numbers = [random.randint(1, 10500) for x in range(10000)]
list_numbers.sort()
print("La list: ")
print(list_numbers)

def binary_search(numbers, start, end, num):
    if end < start:
        return -1
    middle = (start + end) // 2

    if num == numbers[middle]:
        return middle

    elif num < numbers[middle]:
        return binary_search(numbers, start, middle - 1, num)

    else:
        return binary_search(numbers, middle + 1, end, num)


time_start = time.time()

random_num = random.choice(list_numbers)
print(f"Le nombre random est {random_num}")
index = binary_search(list_numbers, 0, len(list_numbers), random_num)

time_end = time.time()
with open("time(dichotomie).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print(f"The index of the number {random_num} is {index}")
```

```python
#Tri par Tas

import time
import random

def heapify(numbers, n, i):
    root = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and numbers[left] > numbers[root]:
        root = left

    if right < n and numbers[right] > numbers[root]:
        root = right

    if root != i:
        numbers[i], numbers[root] = numbers[root], numbers[i]
        heapify(numbers, n, root)

def heap_sort(numbers):
    n = len(numbers)

    for i in range(n // 2 - 1, -1, -1):
        heapify(numbers, n, i)

    for i in range(n - 1, 0, -1):
        numbers[i], numbers[0] = numbers[0], numbers[i]
        heapify(numbers, i, 0)



list_numbers = [random.randint(1, 10500) for x in range(10000)]

print("The list before: ")
print(list_numbers)

time_start = time.time()

heap_sort(list_numbers)
            
time_end = time.time()
with open("time(Tas).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print(f"The total time is {time_end - time_start}\n")
print("The list after: ")
print(list_numbers)
```

```python
#Tri par Fusion

import time
import random

def merge_sort(numbers):
    if len(numbers) <= 1:
        return numbers

    mid = len(numbers) // 2
    left_half = merge_sort(numbers[:mid])
    right_half = merge_sort(numbers[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result



list_numbers = [random.randint(1, 10500) for x in range(10000)]

print("The list before: ")
print(list_numbers)

time_start = time.time()

list_numbers_sorted = merge_sort(list_numbers)
            
time_end = time.time()
with open("time(Fusion).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print(f"The total time is {time_end - time_start}\n")
print("The list after: ")
print(list_numbers_sorted)
```

```python
#Tri par Bulles

import time
import random

list_numbers = [random.randint(1, 10500) for x in range(10000)]

print("The list before: ")
print(list_numbers)

time_start = time.time()

for i in range(len(list_numbers) - 1):
    for j in range(len(list_numbers) - i - 1):
        if list_numbers[j] > list_numbers[j + 1]:
            list_numbers[j], list_numbers[j + 1] = list_numbers[j + 1], list_numbers[j]
            
time_end = time.time()
with open("time(Bulles).txt", "a") as file:
    time_execution = time_end - time_start
    file.write(str(time_execution))
    file.write("\n")

print(f"The total time is {time_end - time_start}\n")
print("The list after: ")
print(list_numbers)
```
